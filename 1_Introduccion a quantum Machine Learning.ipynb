# Para que sirve una computadora cuntica?

En 1981, [Richard Feynmann](https://en.wikipedia.org/wiki/Richard_Feynman) publica un trabajo titulado ["Simulating Physics with Computers"](https://link.springer.com/article/10.1007/BF02650179). Feynmann nota aqu algo tan simple como potente: la fsica es inherentemente cuntica. Simular sistemas fsicos con computadoras tradicionales est fundamentalmente limitado. 
Para entender esta afirmacin, es necesario volver unos aos atrs. Ms en particular, a las discusiones entre Bohr y Einstein.

[Niels Bohr](https://en.wikipedia.org/wiki/Niels_Bohr) lidera espiritualmente el desarrollo de la teora cuntica. Esta teora vena a romper un componente divino de la fsica: el determinismo. La idea de que dadas las condiciones iniciales y conocido el sistema, es posible predecir toda la evolucin futura. En particular, la accin de medir en mecnica cuntica, es capaz de arrojar resultados aleatorios a partir de una distribucin de probabilidad. 
[Einstein](https://en.wikipedia.org/wiki/Albert_Einstein), el ms romntico de los fsicos, se negaba a creer que la fsica no fuera tan divina y perfecta como l la haba imaginado. En 1935 publica un trabajo donde presenta la ["paradoja EPR"](https://en.wikipedia.org/wiki/EPR_paradox), en la cual mostraba que la cuntica concebida de esa forma violara algunas propiedades inviolables de la fsica tradicional. En el acto de medir localmente partculas fuertemente correlacionadas, sera posible transmitir informacin a una velocidad instantanea, rompiendo las leyes de la relatividad que l mismo haba creado. Para l, la explicacin de los comportamientos cunticos deba buscarse en variables ocultas, es decir en fsica determinista ms pequea y desconocida por nosotros. Ese desconomiento clsico sera entonces el origen de la impredecibilidad, y no un componente inherentemente antideterminasta.

Sin embargo, Einstein estaba equivocado. En 1964 Bell publica una serie de [trabajos](https://cds.cern.ch/record/111654/files/vol1p195-200_001.pdf) donde demuestra que los resultados de la mecnica cuntica no podran replicarse con una teora que sea simultaneamente local* y de variables ocultas: si una teora de variables ocultas es local, no es compatible con los resultados de la mecnica cuntica (no es real), y si coincide (es real) entonces es no local .
Si bien este resultado deja la puerta abierta a una teora no local de variables ocultas (ver [teora de Bohm](https://en.wikipedia.org/wiki/De_Broglie%E2%80%93Bohm_theory)) y a otras interpretaciones como la de [Many Worlds](https://en.wikipedia.org/wiki/Many-worlds_interpretation) (donde se abandona la hiptesis de Bell de cada medida tiene un nico resultado), la gran mayoria de la comunidad cree que la cuntica simplemente no es una teora de variables ocultas, y por ende no rompe localidad (y por ende relatividad).

**Volviendo a Feynmann, este resultado implica que las leyes de la teora cuntica no son simulables clsicamente. Para simular sistemas cunticos, es necesario introducir cuntica. De esta forma se empez a pensar en las computadoras cunticas como mquinas con el objetivo de simular sistemas fsicos.**

# Y esto que tiene que ver con Machine Learning?

En los 90, se dieron cuenta de que tratar con sistemas cunticos es, ni mas ni menos, que tratar con lgebra lineal. Una computadora basada en sistemas cunticos tendra entonces una gran ventaja para problemas de este tipo. Se desarrollaron una serie de algoritmos que mostraron ventajas enormes por sobre sus contrapartes clsicos en algunos problemas particulares. Entre estos los ms famosos y que ms llamaron la atencin fueron:

- [Algoritmo de Shor](https://ieeexplore.ieee.org/document/365700): algoritmo que permite factorizar nmeros enteros en su descomposicin en primos. Este problema, que requiere una cantidad de pasos exponencial con la complejidad en computacin clsica, es polinmico en computacin cuntica. Es decir, un problema NP clsicamente deja de serlo con este algoritmo. Este algoritmo puede ser usado para resolver encriptacin RSA por ejemplo, lo que rompera los sistemas de seguridad de bancos y blockchains con proof of work. Podran resolverse en pocos minutos problemas que tardan en promedio millones de aos en computadoras clsicas.

- [Algoritmo de Grover](https://arxiv.org/abs/quant-ph/9605043): busquedas de objetos en lstas desordenadas (ejemplo: buscar un nombre en una guia telefnica desordenada). Requiere $O(\sqrt{N})$ pasos en una computadora cuntica contra $O(N)$ de su contraparte clsico. Es decir, tenemos un speed up proporcional a la raiz cuadrada.


**Como pueden ser tanto ms rpidos?**

La primera intuicin puede ganarse tratando de escribir un vector de largo N. Esto requiere una cantidad $O(N)$ de bits en una computadora clsica, mientras que en su contraparte cuntico requiere $O(log(N))$ qubits. A modo de ejemplo, para representar el vector clsico $v=(1,0,0,1)$, necesitamos 4 bits. El anlogo cuntico necesita solo 2 qubits, ya que con 2 qubits me puedo generar la base ${00,01,10,11}$ cuyos elementos puedo combinar linealmente como $|v\rangle = |00\rangle + |11\rangle = (1,0,0,1)$. 
Entonces solo por empezar, aquellos problemas que escalean con tiempo polinmico $O(N)$, escalaran (en caso de poder traducirse eficientemente) como $O(log(N))$. Algunos ejemplos de subrutinas bsicas de lgebra lineal son:


- [FFT (fast fourier transform)](https://en.wikipedia.org/wiki/Fast_Fourier_transform): Algoritmo para calcular la transformacin discreta de Fourier. Lleva tiempo $N\log N$ en computacin clsica (usando sparcity). El anlogo cuntico lleva tiempo $O((\log N)^2)$. Si algo tarda $10^{14}$ steps, el cuantico lleva tiempo 1600 steps. Pensando en memoria, si $N=10^{12}$, esto equivale a $n=log N = 40 bits$. Mapeo un terabyte a 40 qubits, osea tenemos mucha compresion de memoria. 
- [Hallar autovectores y autovalores de matrices esparsas o de bajo rango](https://arxiv.org/abs/quant-ph/9807070): Clsico lleva $O(N^2)$ pasos, y cuntico $O((\log N)^2)$ pasos. En memoria tengo la misma compresin. Si busco eigensystem de matrices de $N\times N$ con $N=10^{12}$, esto esta completamente fuera de alcanze con $10^{24}$ operaciones. En el anlogo cuntico, lo hago en $1600$ operaciones cunticas.
- [Inversin de Matrices](https://en.wikipedia.org/wiki/Quantum_algorithm_for_linear_systems_of_equations): $Ax=b$, queremos hallar $x=A^{-1}b$. Clsicamente lleva tiempo $O(N^2)$, el anlogo cuntico $A|x\rangle = |b\rangle$ (usamos evolucin natural del sistema, que es invertible) se puede hacer en $O((\log N)^2)$.

Estos son solo algunos de los problemas donde aparecen speed-ups monstruosos, y que son de suma relevancia para Machine Learning. Como se puede ver, tiene sentido el esfuerzo que se hace en esta direccin. Sin embargo hay que aclarar que estamos todava muy lejos de eimplementar este tipo de algoritmos a una escala que resulte til.

Algo fundamental para que todo esto sea posible es conseguir realizar el mapeo de vectores clsicos a estados cunticos. Para esto necesitamos un device llamado qRAM. Actualmente se estn investigando como hacer esto en superconductores, iones, luz. Este es el desafo tcnico donde ms energas se estn volcando en este momento y se espera que sea el prximo gran logro del campo.

# Que problemas de Machine Learning podemos resolver con esto?

Casi todos los algoritmos de Machine Learning dependen fuertemente de operaciones con lgebra lineal. No es de extraar que las ventajas cunticas toquen de alguna manera a practicamente todos los algoritmos. A continuacin menciono solo algunos de ellos, donde queda muy claro en que eslabn aparece el speed up:

### [Cluster Finding](https://en.wikipedia.org/wiki/Cluster_analysis)

Nos dan clusters de data etiquetada, nos dan un nuevo punto y nos preguntan a que cluster pertenece. La idea es calcular el valor medio de cada cluster y medir una distancia euclidea para evaluar pertenencia. Si los puntos viven en dimensiones altas, calcular estas distancias es una tarea bastante dificil. En computadoras cunticas se tiene un speed up notable.

### [Support Vector Machine](https://en.wikipedia.org/wiki/Support_vector_machine)

Tenemos clusters de data, y mi algoritmo quiere construir el hiperplano divisor ptimo.  Esto implica resolver $|b\rangle = A^{-1}|x\rangle$, y por ende tenemos el speed up de inversin de matrices. 

### [Principal Component Analysis](https://en.wikipedia.org/wiki/Principal_component_analysis)

Dado un cluster de data, queremos hallar los ejes que lo definen. Tenemos mejora exponencial. Empresas como Netflix usan este algoritmo para asociar usuarios con gneros y recomendar de forma personalizada. 

### [Topological data analysis](https://en.wikipedia.org/wiki/Topological_data_analysis)

Dado un cluster de data con una topologa extraa (por ejemplo con agujeros), para encontrar las propiedades topolgicas se estudian autovalores y autoestados de matrices enormes. Con computadoras cunticas se tiene entonces un speed up enorme.

### [Deep Learning](https://en.wikipedia.org/wiki/Deep_learning)

Redes complejas de nodos interactuantes, con pesos, de forma tal que ajustando los pesos se logra minimizar una funcin costo y realizar predicciones. Esto se usa para reconocimiento de imgenes, voice recognition, y en general busqueda de patrones en data difciles de obtener con mtodos tradicionales. Lo interesante es que estas redes con parmetros optimizables pueden emularse en computacin cuntica a travs de compuertas lgicas cunticas variacionales. Estas son capaces de encontrar patrones que van ms all de los clsicos. Adem de tener un speed up, tenemos acceso a nuevos patrones.

### Quantum annealing  computacin adiabtica 

[Este tipo de computadoras](https://en.wikipedia.org/wiki/Adiabatic_quantum_computation) apuntan a resolver nicamente problemas de optimizacin. La idea bsica es que toda funcin coste puede mapearse a un Hamiltoniano (operador que describe la energa del sistema), y que no hay algoritmo que minimice ms rpidamente que un sistema fsico evolucionando a su estado fundamental de energa. Este mapeo es universal, es decir que en principio es posible resolver cualquier problema de minimizacin mediante enega. Por supuesto que problemas complejos tienen Hamiltonianos ms complejos y por ende mnimos locales donde nuestro sistema puede trabarse. Para esto se usa el [teorema adiabatico](https://en.wikipedia.org/wiki/Adiabatic_theorem), que dice que si un sistema se encuentra en su estado fundamental, y se deforma el Hamiltoniano lo suficientemente lento, el sistema se queda siempre en el estado fundamental. En este caso se busca evolucionar el sistema con $H=H_0 (1-t) + H_{final} t$, y se usan pasos cortos (cuanto ms corto el paso, ms seguro el algoritmo, pero ms costoso). Tambin puede usarse [quantum annealing](https://en.wikipedia.org/wiki/Adiabatic_theorem), en el cual en lugar de evolucionar lentamente el sistema al Hamiltoniano deseado, se deja minimizar el sistema en el Hamiltoniano final muchas veces partiendo de distintos estados inciales y quedandose finalmente con el estado de menor energa de todas las iteraciones (en realidad se evoluciona la superposicin de todos los posibles candidatos).

Estas computadoras funcionan entonces mapeando problemas de optimizacin a Hamiltonianos de sistemas tipo [Ising](https://en.wikipedia.org/wiki/Ising_model) y haciendolos evolucionar de forma controlada hasta llegar a un estado que a su vez es mapeable a nuevamente como solucin del problema original. Es un optimizador fsico. El speedup obtenido es enorme y dificil de comparar, porque no depende de numero de operaciones, como el gate model, sino del $\Delta t$ usado para evolucionar, o la cantidad de iteraciones en Annealing (que a su vez depende de la complejidad del Hamiltoniano). 

Este mtodo es ms estable frente al ruido. Se usan qubits de flujo, y la empresa que los desarrolla (D-Wave) cuenta con procesadores de hasta 5640 qubits. Se usan para resolver problemas del tipo Max-Cut (hallar los cortes en un grfico que minimicen cierta cantidad), Traveling Salesman (hayar la ruta  entre varios puntos que minimice la distancia total), problemas de Scheduling (hallar distribucin ptima de recursos en el tiempo, con ciertos constraints), y dems cosas de este estilo.

# Diferential programming (quantum deep learning)

La idea es pensar circuitos y algoritmos cunticos como objetos entrenables y obtimizables de la misma forma que una red neuronal. 

En Deep Learning, construimos modelos de Machine Learning con funciones acumulables y diferenciables. La idea es construir una red con parmetros optimizables que pueden "componerse" (acumularse en capas), y a quin se le puede asignar una funcin costo que queremos minimizar. Vamos ajustando los parmetros usando gradient descent (movemos parmetros en direccin del gradiente, es decir hacia los mnimos de la funcin de costo) y finalmente usamos nuestro modelos entrenado para predecir.

Differentiable programming (programacin diferenciable) es un concepto dentro del cual podemos ubicar deep learning. La idea es que todos los cdigos deben ser entrenables, no solo los de machine learning. La idea es que nuestra funcin optimizable puede ser cualquier trozo de cdigo. Es posible definir una funcin costo que dependa de los resultados y usar gradient descent para optimizar. Con esto podemos hacer mucho ms que predecir.

En computacin cuntica, tenemos circuitos con compuertas dependientes de parmetros (rotaciones). Los outputs son valores de expectacin, que son funciones continuas de estos parmetros y por ende diferenciables. Pensar entonces en differentiable programming resulta sumamente natural. 

Cuando hablamos de quantum machine learning, una opcin es usar computadoras cunticas para hacer machine learning usando los speed ups de algebra lineal. La otra es usar herramientas de machine learning (y differentiable programming) para entender mejor los circuitos cunticos. 

[PennyLane](https://pennylane.ai/), [Tensorflow Quantum](https://www.tensorflow.org/quantum), [Tequila](https://github.com/aspuru-guzik-group/tequila) y [Yao](https://yaoquantum.org/) son herramientas opensource que implementan circuitos cunticos como programas diferenciables.

### Como calculamos el gradiente en circuitos cunticos?

- Una opcin es aproximarlo con diferencias finitas. Pero ya hay bastante ruido y aproximaciones en circuitos cunticos.
- Otra opcin es simularlo clsicamente. Esto funciona, solo para circuitos pequeos, sino dea de ser simulable.
- Otra estratega es evitar gradient descent y usar otros mtodos. El tema es que estamos reinventando la rueda y nos perdemos todas las herramientas que conocemos de Machine Learning.

Entonces que suele usarse? Tenemos la regla del Parameter Shift:
$\partial{f(\theta)}/\partial{\theta}=c[f(\theta+s)-f(\theta-s)]$
Esto no vale en general, pero si para algunas funciones. En particular, es vlido para aquellas de estructura trigonomtrica. Como la estructura de  nmeros complejas coincide con la trigonomtrica en muchos aspectos, los circuitos cunticos admiten este tipo de descomposicin de forma exacta. 
Esta regla es exacta para este tipo de funciones, no es una aproximacin. Es distinto a aproximar la derivada con un $\Delta \theta$ finito (diferencias finitas), que s es una aproximacin y  requiere shifts muy pequeos. En este caso el shift es de tamao arbitrario. El mtodo de diferencias finitas, lo que tienen de malo es que introducen nuevas aproximaciones que preferimos evitar.

El parameter shift lo podemos usar tanto en hardware como en simuladores. No todas las compuertas admiten parameter shift, pero la grn mayora s. Cada compuerta tiene su regla, pero es siempre igual para cada tipo de compuerta (no depende del contexto).

A la hora de resolver un problema en una computadora cuntica, vamos a definir un ansatz (una condicin inicial), una funcin costo, y luego vamos a entrenar evaluando nuestra funcin costo y moviendo los parmetros en la direccin del gradiente.

 El algoritmo ms famoso de este tipo es el Variational Quantum Eigensolver (VQE), que se usa en problemas qumicos. En este caso, el ansatz es algun estado cuntico que aproxima al estado exacto de la molcula, la funcin costo est relacionada con la energa de la molcula, y para optimizar se usa gradient descent. Este algoritmo va a buscar el estado de menor energa para este sistema.

Otro algoritmo muy importante de este tipo es Quatnum Approximate Optimization Algorithm (QAOA), que busca la solucin de un conjunto finito de posibles soluciones para algn problema complejo con constraints. El ansatz depende del problema, pero suele ser bastante complejo (evoluciones repetidas con unitarias), la funcin costo tiene que ver con la energa de un problema equivalente de tipo Ising (ver notebooks siguientes) y se optimiza tambin con gradient descent. 

Algunos algoritmos importantes de este tipo son:

- Problemas de fsica y qumica: Variational Quantum Eigensolver, Variational Quantum Thermalizer, Variational Quantum Diagonalization, Variational Quantum Hamiltonian Diagonalization.
- Problemas de matemtica: Variational Quantum Factoring, Variational Quantum Linear Solver
- Machine Learning: Quantum GANs, Quantum Classifier, Quantum Optical Neural Networks, Quantum Graph Recurrent Networks, Quanvolutional Neural Networks, Quantum Convolutional Neural Networks.

Estrategias de differentiable programming se usan tambien para hacer embedding de data en los circuitos. Es decir, agregar una capa entrenable de embedding.
