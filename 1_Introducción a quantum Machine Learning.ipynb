# ¿Para que sirve una computadora cuántica?

En 1981, [Richard Feynmann](https://en.wikipedia.org/wiki/Richard_Feynman) publica un trabajo titulado ["Simulating Physics with Computers"](https://link.springer.com/article/10.1007/BF02650179). Feynmann nota aquí algo tan simple como potente: la física es inherentemente cuántica. Simular sistemas físicos con computadoras tradicionales está fundamentalmente limitado. 
Para entender esta afirmación, es necesario volver unos años atrás. Más en particular, a las discusiones entre Bohr y Einstein.

paperfeyn.png



[Niels Bohr](https://en.wikipedia.org/wiki/Niels_Bohr) lidera espiritualmente el desarrollo de la teoría cuántica. Esta teoría venía a romper un componente divino de la física: el determinismo. La idea de que dadas las condiciones iniciales y conocido el sistema, es posible predecir toda la evolución futura. En particular, la acción de medir en mecánica cuántica, es capaz de arrojar resultados aleatorios a partir de una distribución de probabilidad. 
[Einstein](https://en.wikipedia.org/wiki/Albert_Einstein), el más romántico de los físicos, se negaba a creer que la física no fuera tan divina y perfecta como él la había imaginado. En 1935 publica un trabajo donde presenta la ["paradoja EPR"](https://en.wikipedia.org/wiki/EPR_paradox), en la cual mostraba que la cuántica concebida de esa forma violaría algunas propiedades inviolables de la física tradicional. En el acto de medir localmente partículas fuertemente correlacionadas, sería posible transmitir información a una velocidad instantanea, rompiendo las leyes de la relatividad que él mismo había creado. Para él, la explicación de los comportamientos cuánticos debía buscarse en variables ocultas, es decir en física determinista más pequeña y desconocida por nosotros. Ese desconomiento clásico sería entonces el origen de la impredecibilidad, y no un componente inherentemente antideterminasta.

Sin embargo, Einstein estaba equivocado. En 1964 Bell publica una serie de [trabajos](https://cds.cern.ch/record/111654/files/vol1p195-200_001.pdf) donde demuestra que los resultados de la mecánica cuántica no podrían replicarse con una teoría que sea simultaneamente local* y de variables ocultas: “si una teoría de variables ocultas es local, no es compatible con los resultados de la mecánica cuántica (no es real), y si coincide (es real) entonces es no local” .
Si bien este resultado deja la puerta abierta a una teoría no local de variables ocultas (ver [teoría de Bohm](https://en.wikipedia.org/wiki/De_Broglie%E2%80%93Bohm_theory)) y a otras interpretaciones como la de [Many Worlds](https://en.wikipedia.org/wiki/Many-worlds_interpretation) (donde se abandona la hipótesis de Bell de cada medida tiene un único resultado), la gran mayoria de la comunidad cree que la cuántica simplemente no es una teoría de variables ocultas, y por ende no rompe localidad (y por ende relatividad).

**Volviendo a Feynmann, este resultado implica que las leyes de la teoría cuántica no son simulables clásicamente. Para simular sistemas cuánticos, es necesario introducir cuántica. De esta forma se empezó a pensar en las computadoras cuánticas como máquinas con el objetivo de simular sistemas físicos.**

# ¿Y esto que tiene que ver con Machine Learning?

En los 90, se dieron cuenta de que tratar con sistemas cuánticos es, ni mas ni menos, que tratar con álgebra lineal. Una computadora basada en sistemas cuánticos tendría entonces una gran ventaja para problemas de este tipo. Se desarrollaron una serie de algoritmos que mostraron ventajas enormes por sobre sus contrapartes clásicos en algunos problemas particulares. Entre estos los más famosos y que más llamaron la atención fueron:

- [Algoritmo de Shor](https://ieeexplore.ieee.org/document/365700): algoritmo que permite factorizar números enteros en su descomposición en primos. Este problema, que requiere una cantidad de pasos exponencial con la complejidad en computación clásica, es polinómico en computación cuántica. Es decir, un problema NP clásicamente deja de serlo con este algoritmo. Este algoritmo puede ser usado para resolver encriptación RSA por ejemplo, lo que rompería los sistemas de seguridad de bancos y blockchains con proof of work. Podrían resolverse en pocos minutos problemas que tardan en promedio millones de años en computadoras clásicas.

- [Algoritmo de Grover](https://arxiv.org/abs/quant-ph/9605043): busquedas de objetos en lístas desordenadas (ejemplo: buscar un nombre en una guia telefónica desordenada). Requiere $O(\sqrt{N})$ pasos en una computadora cuántica contra $O(N)$ de su contraparte clásico. Es decir, tenemos un speed up proporcional a la raiz cuadrada.


**Como pueden ser tanto más rápidos?**

La primera intuición puede ganarse tratando de escribir un vector de largo N. Esto requiere una cantidad $O(N)$ de bits en una computadora clásica, mientras que en su contraparte cuántico requiere $O(log(N))$ qubits. A modo de ejemplo, para representar el vector clásico $v=(1,0,0,1)$, necesitamos 4 bits. El análogo cuántico necesita solo 2 qubits, ya que con 2 qubits me puedo generar la base ${00,01,10,11}$ cuyos elementos puedo combinar linealmente como $|v\rangle = |00\rangle + |11\rangle = (1,0,0,1)$. 
Entonces solo por empezar, aquellos problemas que escalean con tiempo polinómico $O(N)$, escalarían (en caso de poder traducirse eficientemente) como $O(log(N))$. Algunos ejemplos de subrutinas básicas de álgebra lineal son:


- [FFT (fast fourier transform)](https://en.wikipedia.org/wiki/Fast_Fourier_transform): Algoritmo para calcular la transformación discreta de Fourier. Lleva tiempo $N\log N$ en computación clásica (usando sparcity). El análogo cuántico lleva tiempo $O((\log N)^2)$. Si algo tarda $10^{14}$ steps, el cuantico lleva tiempo 1600 steps. Pensando en memoria, si $N=10^{12}$, esto equivale a $n=log N = 40 bits$. Mapeo un terabyte a 40 qubits, osea tenemos mucha compresion de memoria. 
- [Hallar autovectores y autovalores de matrices esparsas o de bajo rango](https://arxiv.org/abs/quant-ph/9807070): Clásico lleva $O(N^2)$ pasos, y cuántico $O((\log N)^2)$ pasos. En memoria tengo la misma compresión. Si busco eigensystem de matrices de $N\times N$ con $N=10^{12}$, esto esta completamente fuera de alcanze con $10^{24}$ operaciones. En el análogo cuántico, lo hago en $1600$ operaciones cuánticas.
- [Inversión de Matrices](https://en.wikipedia.org/wiki/Quantum_algorithm_for_linear_systems_of_equations): $Ax=b$, queremos hallar $x=A^{-1}b$. Clásicamente lleva tiempo $O(N^2)$, el análogo cuántico $A|x\rangle = |b\rangle$ (usamos evolución natural del sistema, que es invertible) se puede hacer en $O((\log N)^2)$.

Estos son solo algunos de los problemas donde aparecen speed-ups monstruosos, y que son de suma relevancia para Machine Learning. Como se puede ver, tiene sentido el esfuerzo que se hace en esta dirección. Sin embargo hay que aclarar que estamos todavía muy lejos de eimplementar este tipo de algoritmos a una escala que resulte útil.

Algo fundamental para que todo esto sea posible es conseguir realizar el mapeo de vectores clásicos a estados cuánticos. Para esto necesitamos un device llamado qRAM. Actualmente se están investigando como hacer esto en superconductores, iones, luz. Este es el desafío técnico donde más energías se están volcando en este momento y se espera que sea el próximo gran logro del campo.

# Que problemas de Machine Learning podemos resolver con esto?

Casi todos los algoritmos de Machine Learning dependen fuertemente de operaciones con álgebra lineal. No es de extrañar que las ventajas cuánticas toquen de alguna manera a practicamente todos los algoritmos. A continuación menciono solo algunos de ellos, donde queda muy claro en que eslabón aparece el speed up:

### [Cluster Finding](https://en.wikipedia.org/wiki/Cluster_analysis)

Nos dan clusters de data etiquetada, nos dan un nuevo punto y nos preguntan a que cluster pertenece. La idea es calcular el valor medio de cada cluster y medir una distancia euclidea para evaluar pertenencia. Si los puntos viven en dimensiones altas, calcular estas distancias es una tarea bastante dificil. En computadoras cuánticas se tiene un speed up notable.

### [Support Vector Machine](https://en.wikipedia.org/wiki/Support_vector_machine)

Tenemos clusters de data, y mi algoritmo quiere construir el hiperplano divisor óptimo.  Esto implica resolver $|b\rangle = A^{-1}|x\rangle$, y por ende tenemos el speed up de inversión de matrices. 

### [Principal Component Analysis](https://en.wikipedia.org/wiki/Principal_component_analysis)

Dado un cluster de data, queremos hallar los ejes que lo definen. Tenemos mejora exponencial. Empresas como Netflix usan este algoritmo para asociar usuarios con géneros y recomendar de forma personalizada. 

### [Topological data analysis](https://en.wikipedia.org/wiki/Topological_data_analysis)

Dado un cluster de data con una topología extraña (por ejemplo con agujeros), para encontrar las propiedades topológicas se estudian autovalores y autoestados de matrices enormes. Con computadoras cuánticas se tiene entonces un speed up enorme.

### [Deep Learning](https://en.wikipedia.org/wiki/Deep_learning)

Redes complejas de nodos interactuantes, con pesos, de forma tal que ajustando los pesos se logra minimizar una función costo y realizar predicciones. Esto se usa para reconocimiento de imágenes, voice recognition, y en general busqueda de patrones en data difíciles de obtener con métodos tradicionales. Lo interesante es que estas redes con parámetros optimizables pueden emularse en computación cuántica a través de compuertas lógicas cuánticas variacionales. Estas son capaces de encontrar patrones que van más allá de los clásicos. Ademá de tener un speed up, tenemos acceso a nuevos patrones.

# Donde estamos ahora?

Sinceramente estamos lejos de tener computadoras cuánticas capaces de ejecutar todos estos algoritmos de machine learning a una escala "útil". Para tener una idea, hoy las computadoras de gates (modelo más popular) más potentes no alcanzan lo 100 qubits.  Para crackear encriptacion 2048-RSA con el algoritmo de Shor, se necesitan contar con alrededor de 4000 logical qubits. Cada logical quibt tiene alrededor de 1000 qubits reales. Es decir necesitamos 4 millones de qubits para este problema. 

## ¿Que están haciendo hoy por hoy las computadoras cuánticas?
Para responder esto tenemos que distinguir entre las dos grandes familias de computadoras cuánticas existentes (hay más, pero no son las opciones principales en este momento).

### Gate model

La idea de estos dispositivos es emular las compuertas lógicas  usadas en computación tradicional. Sin embargo, estas compuertas son ahora operaciones cuánticas, tales como rotar, o hacer cambios controlados. Para ello se alimenta el sistema con qubits (generalmente electrones), que se encuentran en el estado incial $|0\rangle$, se les aplica una serie de operaciones  y se miden las propiedades de estas partículas. 

La computadora más potente de este tipo cuenta en la actualidad con 65 qubits, pero se espera que este número crezca rápidamente en los próximos años (IBM estima 1000 para 2023). Tiene algunas propiedades muy interesantes. Por ejemplo, mientras que en computación clásica las compuertas de 1 bit son solo la identidad y la negación, ahora tenemos tanto la identidad como el conjunto de rotaciones en el espacio complejo (las llamadas [operaciones de Pauli](https://en.wikipedia.org/wiki/Pauli_matrices), donde $\sigma_x$ equivale a la negación y las otras son rotaciones con fases). Además, es posible conseguir computación universal con unas pocas compuertas, es decir que en teoría podemos generar cualquier estado cuántico a partir de un conjunto reducido de operaciones. La dificultad está en saber que compuertas tomar para llegar al estado que queremos. Y este suele ser el objetivo de los algoritmos cuánticos (Shor, Grover y Deutsch Jozsa caen en esta categoría de gate-model). 

La gran desventaja de este método es el ruido. Es muy dificil mantener un estado cuántico, y para ello suele empearse baja temperatura, y una serie de algoritmos cuánticos de correción de errores. Es por esto que cada qubit lógico, está en realidad hecho por varios qubits fisicos que actúan de reductores de ruido.

En la industria se están usando este tipo de computadoras en conjunto con computadoras clásicas (métodos híbridos). Se usan por ejemplo en problemas moleculares o que requieran simular sistemas genuinamente cuánticos. El Variational Quantum Eigensolvers (VQE) es sin duda el más estudiado. El objetivo de VQE es hallar mínimos de ciertas cantidades. En partícular, es muy usado para hallar las energías y los estados fundamentales de ciertas [moléculas](https://qiskit.org/textbook/ch-applications/vqe-molecules.html) o problemas de optimizacion del tipo [Max-Cut](https://medium.com/qiskit/the-variational-quantum-eigensolver-43f7718c2747). La idea es usar circuitos con parámetros optimizables (ángulos de rotación), medir ciertos observables al final de todo, pasar la información a una computadora clásica que haga algún tipo de gradient descent para optimizar los ángulos y repetir. De esta forma se van alcanzando los ángulos de rotación ideales para minimizar alguna función de costo.

### Quantum annealing ó computación adiabática 

[Este tipo de computadoras](https://en.wikipedia.org/wiki/Adiabatic_quantum_computation) apuntan a resolver únicamente problemas de optimización. La idea básica es que toda función coste puede mapearse a un Hamiltoniano (operador que describe la energía del sistema), y que no hay algoritmo que minimice más rápidamente que un sistema físico evolucionando a su estado fundamental de energía. Este mapeo es universal, es decir que en principio es posible resolver cualquier problema de minimización mediante enegía. Por supuesto que problemas complejos tienen Hamiltonianos más complejos y por ende mínimos locales donde nuestro sistema puede trabarse. Para esto se usa el [teorema adiabatico](https://en.wikipedia.org/wiki/Adiabatic_theorem), que dice que si un sistema se encuentra en su estado fundamental, y se deforma el Hamiltoniano lo suficientemente lento, el sistema se queda siempre en el estado fundamental. En este caso se busca evolucionar el sistema con $H=H_0 (1-t) + H_{final} t$, y se usan pasos cortos (cuanto más corto el paso, más seguro el algoritmo, pero más costoso). También puede usarse [quantum annealing](https://en.wikipedia.org/wiki/Adiabatic_theorem), en el cual en lugar de evolucionar lentamente el sistema al Hamiltoniano deseado, se deja minimizar el sistema en el Hamiltoniano final muchas veces partiendo de distintos estados inciales y quedandose finalmente con el estado de menor energía de todas las iteraciones (en realidad se evoluciona la superposición de todos los posibles candidatos).

Estas computadoras funcionan entonces mapeando problemas de optimización a Hamiltonianos de sistemas tipo [Ising](https://en.wikipedia.org/wiki/Ising_model) y haciendolos evolucionar de forma controlada hasta llegar a un estado que a su vez es mapeable a nuevamente como solución del problema original. Es un optimizador físico. El speedup obtenido es enorme y dificil de comparar, porque no depende de numero de operaciones, como el gate model, sino del $\Delta t$ usado para evolucionar, o la cantidad de iteraciones en Annealing (que a su vez depende de la complejidad del Hamiltoniano). 

Este método es más estable frente al ruido. Se usan qubits de flujo, y la empresa que los desarrolla (D-Wave) cuenta con procesadores de hasta 5640 qubits. Se usan para resolver problemas del tipo Max-Cut (hallar los cortes en un gráfico que minimicen cierta cantidad), Traveling Salesman (hayar la ruta  entre varios puntos que minimice la distancia total), problemas de Scheduling (hallar distribución óptima de recursos en el tiempo, con ciertos constraints), y demás cosas de este estilo.

# Diferential programming (quantum deep learning)

La idea es pensar circuitos y algoritmos cuánticos como objetos entrenables y obtimizables de la misma forma que una red neuronal. 

En Deep Learning, construimos modelos de Machine Learning con funciones acumulables y diferenciables. La idea es construir una red con parámetros optimizables que pueden "componerse" (acumularse en capas), y a quién se le puede asignar una función costo que queremos minimizar. Vamos ajustando los parámetros usando gradient descent (movemos parámetros en dirección del gradiente, es decir hacia los mínimos de la función de costo) y finalmente usamos nuestro modelos entrenado para predecir.

Differentiable programming (programación diferenciable) es un concepto dentro del cual podemos ubicar deep learning. La idea es que todos los códigos deben ser entrenables, no solo los de machine learning. La idea es que nuestra función optimizable puede ser cualquier trozo de código. Es posible definir una función costo que dependa de los resultados y usar gradient descent para optimizar. Con esto podemos hacer mucho más que predecir.

En computación cuántica, tenemos circuitos con compuertas dependientes de parámetros (rotaciones). Los outputs son valores de expectación, que son funciones continuas de estos parámetros y por ende diferenciables. Pensar entonces en differentiable programming resulta sumamente natural. 

Cuando hablamos de quantum machine learning, una opción es usar computadoras cuánticas para hacer machine learning usando los speed ups de algebra lineal. La otra es usar herramientas de machine learning (y differentiable programming) para entender mejor los circuitos cuánticos. 

[PennyLane](https://pennylane.ai/), [Tensorflow Quantum](https://www.tensorflow.org/quantum), [Tequila](https://github.com/aspuru-guzik-group/tequila) y [Yao](https://yaoquantum.org/) son herramientas opensource que implementan circuitos cuánticos como programas diferenciables.

### Como calculamos el gradiente en circuitos cuánticos?

- Una opción es aproximarlo con diferencias finitas. Pero ya hay bastante ruido y aproximaciones en circuitos cuánticos.
- Otra opción es simularlo clásicamente. Esto funciona, solo para circuitos pequeños, sino dea de ser simulable.
- Otra estrategía es evitar gradient descent y usar otros métodos. El tema es que estamos reinventando la rueda y nos perdemos todas las herramientas que conocemos de Machine Learning.

Entonces que suele usarse? Tenemos la regla del Parameter Shift:
$\partial{f(\theta)}/\partial{\theta}=c[f(\theta+s)-f(\theta-s)]$
Esto no vale en general, pero si para algunas funciones. En particular, es válido para aquellas de estructura trigonométrica. Como la estructura de  números complejas coincide con la trigonométrica en muchos aspectos, los circuitos cuánticos admiten este tipo de descomposición de forma exacta. 
Esta regla es exacta para este tipo de funciones, no es una aproximación. Es distinto a aproximar la derivada con un $\Delta \theta$ finito (diferencias finitas), que sí es una aproximación y  requiere shifts muy pequeños. En este caso el shift es de tamaño arbitrario. El método de diferencias finitas, lo que tienen de malo es que introducen nuevas aproximaciones que preferimos evitar.

El parameter shift lo podemos usar tanto en hardware como en simuladores. No todas las compuertas admiten parameter shift, pero la grán mayoría sí. Cada compuerta tiene su regla, pero es siempre igual para cada tipo de compuerta (no depende del contexto).

A la hora de resolver un problema en una computadora cuántica, vamos a definir un ansatz (una condición inicial), una función costo, y luego vamos a entrenar evaluando nuestra función costo y moviendo los parámetros en la dirección del gradiente.

 El algoritmo más famoso de este tipo es el Variational Quantum Eigensolver (VQE), que se usa en problemas químicos. En este caso, el ansatz es algun estado cuántico que aproxima al estado exacto de la molécula, la función costo está relacionada con la energía de la molécula, y para optimizar se usa gradient descent. Este algoritmo va a buscar el estado de menor energía para este sistema.

Otro algoritmo muy importante de este tipo es Quatnum Approximate Optimization Algorithm (QAOA), que busca la solución de un conjunto finito de posibles soluciones para algún problema complejo con constraints. El ansatz depende del problema, pero suele ser bastante complejo (evoluciones repetidas con unitarias), la función costo tiene que ver con la energía de un problema equivalente de tipo Ising (ver notebooks siguientes) y se optimiza también con gradient descent. 

Algunos algoritmos importantes de este tipo son:

- Problemas de física y química: Variational Quantum Eigensolver, Variational Quantum Thermalizer, Variational Quantum Diagonalization, Variational Quantum Hamiltonian Diagonalization.
- Problemas de matemática: Variational Quantum Factoring, Variational Quantum Linear Solver
- Machine Learning: Quantum GANs, Quantum Classifier, Quantum Optical Neural Networks, Quantum Graph Recurrent Networks, Quanvolutional Neural Networks, Quantum Convolutional Neural Networks.

Estrategias de differentiable programming se usan tambien para hacer embedding de data en los circuitos. Es decir, agregar una capa entrenable de embedding.
